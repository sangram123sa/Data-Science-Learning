```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(-10, 10, 100)
y = x**2
```

```python
plt.plot(x, y)
plt.show()
```

```python
x1 = np.linspace(-10, 10, 100)
x2 = np.linspace(-10, 10, 100)

xx1, xx2 = np.meshgrid(x1, x2)

y = xx1**2 + xx2**2

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(xx1, xx2, y)
# plt.show()
plt.show()
```

```python
x1 = np.linspace(-10, 10, 100)
x2 = np.linspace(-10, 10, 100)

xx1, xx2 = np.meshgrid(x1, x2)

y = xx1**2 - xx2**2

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(xx1, xx2, y)
# plt.show()
plt.show()
```
