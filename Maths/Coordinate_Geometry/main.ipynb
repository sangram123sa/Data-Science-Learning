```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
toy_data = np.array([
    [0.91, 1.13, 1],
    [0.71, 0.59, 0],
    [1.37, 0.23, 1],
    [1.21, 0.72, 1],
    [0.36, 1.44, 0],
    [0.99, 0.87, 1]
])
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x = toy_data[:,0],y=toy_data[:,1], hue=toy_data[:,2] )
plt.xlabel("length")
plt.ylabel("width")
plt.title("Toy Data")
plt.legend()
plt.show()
```

```python
x = np.linspace(start=-5, stop=15, num=100)
y1 = 5/2*x - 3
y2 = -7/3*x +2
```

```python
plt.plot(x, y1, c='blue', label='y=5/2x - 3' )
plt.plot(x, y2, c='red', label='-7/3x+2')
plt.grid()
plt.legend()
plt.show()
```

```python
fig = plt.figure()
# axis add
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-5, 5, 101)
y = np.linspace(-5, 5, 101)

xx, yy = np.meshgrid(x, y)

z = (4/2)*xx -(5/2)*yy

ax.plot_surface(xx, yy,z, alpha = 0.5)

ax.scatter(0,0,0, color = "green")

plt.show()
```
