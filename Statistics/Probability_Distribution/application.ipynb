```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
```

```python
stats.uniform.rvs?
```

```python
# here rvs means random variates
# Generate random data from a uniform distribution
# with a range of 0 to 10
# and a size of 100
uniform_distribution_data = stats.uniform.rvs(loc = 1, scale = 10, size = 100)
```

```python
# We created a histogram of the uniform distribution data
sns.histplot(uniform_distribution_data, bins=10)
plt.show()
```

```python
sns.distplot(uniform_distribution_data, bins=10)
```

```python
stats.uniform.pdf(20, loc = 0, scale = 100)
# The probability density function (PDF) of a uniform distribution
```

```python
stats.uniform.cdf(11, loc = 0, scale = 100)
# The cumulative distribution function (CDF) of a uniform distribution
```

```python
x = np.arange(-40, 140, 1)
```

```python
cdf_uniform = stats.uniform.cdf(x, loc=1, scale=100)
cdf_uniform
```

```python
# ploting the CDF
plt.plot(cdf_uniform)
```

```python
norm = stats.norm(loc = 70, scale = 5)
```

```python
norm.rvs()
```

```python
norm.rvs(size = 1000)
# Generate random data from a normal distribution
```

```python
marks = norm.rvs(size = 10000).astype(int)
marks.shape
# Generate random data from a normal distribution
```

```python
sns.histplot(marks, bins=100,stat='density')
plt.show()
```

```python
sns.distplot(marks, bins=100)
plt.show()
```
