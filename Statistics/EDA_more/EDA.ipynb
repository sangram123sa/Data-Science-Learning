```python
movie = pd.read_csv('movies_metadata.csv')
```

```python
movie.head()
```

```python
movie.info()
```

```python
mov2 = movie[['vote_average', 'vote_count', 'release_date']]
```

```python
mov2 = mov2.dropna()
```

```python
mov2.info()
```

```python
mov2['date'] = pd.to_datetime(mov2['release_date'])
```

```python
mov2.head()
```

```python
mov2['year'] = mov2.date.dt.year
```

```python
mov2.index = mov2['date']
```

```python
mov2.head()
```

```python
fig, ax = plt.subplots()
ax = mov2.vote_average.plot(label='Average rating by year')
```

```python
sns.boxplot(x='year', y='vote_average', data=mov2)
```

```python
mov3 = mov2[mov2['year'] > 1990]
```

```python
plt.figure(figsize=(16, 8))
sns.boxplot(x='year', y='vote_average', data=mov3, hue='year')
```

```python
fig, ax = plt.subplots()

ax = mov3.vote_average.plot(label='Average rating by year', ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(16, 8))  # Set the figure size
ax = mov3.vote_average.plot(label='Average rating by year', ax=ax)
ax.set_xlabel('Year')
ax.set_ylabel('Average Rating')
ax.set_title('Average Movie Rating by Year')
ax.legend()
```

```python
# black and white movies are good or colored movies are good
mov4 = mov2[(mov2['year'] > 1970) & (mov2['year'] < 1985)]
```

```python
plt.figure(figsize=(16, 8))
sns.boxplot(x='year', y='vote_average', data=mov4, hue='year')
```
